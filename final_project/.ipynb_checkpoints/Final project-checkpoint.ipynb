### Выбор модели 

Так как данных достаточно мало, но тем не менее в каждом из рядов может меняться тренд и присутствует недельная сезонность - в качестве прогностической модели была выбрана модель тройного экспоненциального сглаживания (иногда в процессе симуляции чередовалась с моделью двойного экспоненциального сглаживания, где было недостаточно точек для сбора данных о сезонности).    


Доверительный интервал прогноза, насколько его можно назвать таковым, рассчитывался на основании дисперсии остатков модели (*y_hat+-1.96\*SE*, либо же *y_hat+-2.17\*SE*, в зависимости от того, какой интвервал мы хотим построить, где *SE = np.std(residuals)*).   


### Входящие данные

- **start_index** - с какого дня начинаем симуляцию модели;  
- **rolling_window** - окно, за промежуток которого мы строим модель (желательно, чтобы было < start_index);   
- **trigger_window** - сколько дней к ряду должны наблюдаться аномалии, до тех пор пока мы о ней уведомим;  
- **recovery_window** - если к-во аномалий к ряду достигает recovery_window, то предыдущие точки, на основе которых мы делали прогноз удаляются из массива и мы перестраиваем модель на основе последних точек (к-во новых точек = recovery_window). Так как recovery_window = 7, и *seasonal_periods* в нашей модели также = 7, то 7 дней после такого сброса модели - используется модель двойного экспоненциального сглаживания, вместо тройного. 

Все вышеуказанные параметры можно изменять и смотреть что получится :) 

### Описание алгоритма


Алгоритм делает прогноз на 1 день, строит доверительный интервал относительно точки прогноза и на следующий день мы проверяем, попадает ли наша точка в доверительный интервал. 

- **Если попадает** - мы добавляем точку в список, на основе которого мы тренируем модель (переменная *train_s*) и строим новый прогноз (за последние *rolling_window* дней). 

- **Если не попадает и при этом к-во аномалий в сумме к ряду < trigger_window** - мы, формально, не делаем уведомления (хотя в коде print об этом я выводил). Также, мы можем принять такую ситуацию как "предупреждение". Такая точка в список train_s не попадает и мы делаем прогноз на день k+1, где k - это к-во аномалий к ряду; 

- **Если к-во аномалий к ряду == trigger_window** - выводим уведомление, алгоритм действий такой же как в предыдущем подпункте; 

- **Если к-во аномалий к ряду == recovery_window** - сбрасываем наши предыдущие тренировочные данные, перестаем считать последние точки (из recovery_window) как аномалии и берем их как новые тренировочные данные. 


### Плюсы и минусы алгоритма / возможности доработки

Плюсы:

> \+ Алгоритм достаточно гибкий и адаптивный

> \+ Учитывает тренд и недельную сезонность трафика 

Минусы: 

> \-- Сам код не выстроен таким образом, чтобы проверка происходила одновременно по всем рядам. Если это учесть, уведомление об аномалии могло бы звучать более четко и конкретизировано. Например, "Аномальный рост трафик на home page и падение на search page -> падение конверсии в поиск". 

> \-- Подход с доверительными интервалами слишком "чувствительный" -> достаточно много точек выбрасываются из тренировочного сета как "аномалии", хотя не всегда таковыми являются. 

> \-- В текущей реализации нет выявления аномалий на последнем этапе воронки. Хотел бы уточнить, что данную модель я бы там в любом случае не применял, так как имеющиеся числа на этом уровне очень низкие и "случайные". Скорее всего, здесь стоило бы делать какой-то rule-based подход. Например, за аномалию определять дни со значениями выше выставленного порога и дни когда конверсия = 0%. 

> \-- Как было указано выше - модель ничего не говорит об изменениях непосредственно в конверсии, анализ аномалий разделен по каждому этапу воронки и нет какой-то "связности" или иерархии, однако, у меня пока что нет идей как такую "связность" можно было бы реализовать. 


## Model build

```python
def plot_results(result, page): 
    fig, ax = plt.subplots()
    
    data= result.copy()
    
    anomaly_index = data[data.anomaly == 1].date.values
    anomaly_values = data[data.anomaly == 1].real.values
    normal_index = data[data.anomaly != 1].date.values
    normal_values = data[data.anomaly != 1].real.values

    data.set_index('date', inplace=True)

    ax.xaxis.set_major_locator(mdates.WeekdayLocator())
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
    ax.tick_params(axis='x',rotation=45)

    ax.scatter(anomaly_index, anomaly_values, s = 100, c='r', marker="^", label = 'Anomaly real values')
    ax.scatter(normal_index, normal_values, color='black', label = 'Normal real values')
    ax.plot(data.index, data.predict, linewidth = 2,color='blue', label = 'Predicted values')
    ax.fill_between(data.index,
                   data['lower_bound'],
                   data['upper_bound'],
                   color = 'slategray',
                   alpha = 0.3)

    ax.set_title(f'Simulation results, {page} page')
    ax.legend()
    plt.show()
```

```python
def modelling_anomaly_detection(conv_rates_total, page, start_index, 
                               rolling_window, trigger_window, recovery_window,
                               conf_int): 

    model_seq = conv_rates_total[[page]].reset_index().copy()

#     start_index = 14
#     rolling_window = 14
#     trigger_window = 3 
#     recovery_window = 7 
#     conf_int = 1.96

    train_s = model_seq.loc[:start_index-1][page].values
    test_s = model_seq.loc[start_index:][page].values


    model_seq.rename({page:'real'},axis=1,inplace=True)
    model_seq['anomaly'] = np.nan 
    model_seq['lower_bound'] = np.nan
    model_seq['upper_bound'] = np.nan
    model_seq['predict'] = np.nan
    model_seq['resid'] = np.nan
    model_seq = model_seq.reset_index()


    model = ExponentialSmoothing(train_s, trend='add',
                                 seasonal='add',
                                 seasonal_periods=7,
                                 damped_trend=0.3).fit()

    predict = model.fittedvalues
    resids = predict - train_s


    model_seq.loc[:start_index-1, 'predict'] = predict
    model_seq.loc[:start_index-1, 'lower_bound'] = model_seq.loc[:start_index-1, 
                                                                 'predict'] - (conf_int * np.std(resids))
    model_seq.loc[:start_index-1, 'upper_bound'] = model_seq.loc[:start_index-1, 
                                                                 'predict'] + (conf_int * np.std(resids))
    model_seq.loc[:start_index-1, 'resid'] = resids

    predict_1day = model.forecast(1)[0]
    model_seq.loc[start_index, 'predict'] = predict_1day
    model_seq.loc[start_index, 'lower_bound'] = predict_1day - (conf_int * np.std(resids))
    model_seq.loc[start_index, 'upper_bound'] = predict_1day + (conf_int * np.std(resids))


    anomalies_in_a_row = 0

    for i in range(start_index, start_index + len(test_s)): 
        observed = model_seq.loc[i,'real']
        predicted = model_seq.loc[i,'predict']

        model_seq.loc[i, 'resid'] = predicted - observed

        if observed > model_seq.loc[i, 'upper_bound'] or observed < model_seq.loc[i,'lower_bound']:
            
            print('Current date =', model_seq.loc[i,'date'].strftime('%Y-%m-%d'))
            model_seq.loc[i, 'anomaly'] = 1
            anomalies_in_a_row += 1

            if model_seq.loc[i - recovery_window + 1:i, 'anomaly'].sum() == recovery_window: 
                print('>>>>> Recovery window is exceeded, adapting the model <<<<<<')

                anomalies_in_a_row = 0 
                train_s = model_seq.loc[i - recovery_window + 1:i,'real'].values
                #model = ExponentialSmoothing(train_s[-rolling_window:], trend='add',
                #                                     damped_trend=0.2).fit()

                model = SimpleExpSmoothing(train_s[-rolling_window:]).fit()

                predict = model.fittedvalues
                resids = predict - train_s

                predict_1day = model.forecast(1)[0]

                model_seq.loc[i+1, 'predict'] = predict_1day
                model_seq.loc[i+1, 'lower_bound'] = predict_1day - (conf_int * np.std(resids[:-rolling_window]))
                model_seq.loc[i+1, 'upper_bound'] = predict_1day + (conf_int * np.std(resids[:-rolling_window]))



            elif (anomalies_in_a_row < recovery_window) and (anomalies_in_a_row >= trigger_window): 
                print(' >>>>> Anomaly detected, trigger IS REACHED! <<<<<<')
                print(f'Anomalies in a row = {anomalies_in_a_row}')
                ### add plot


                predict_1day = model.forecast(anomalies_in_a_row + 1)[-1]
                model_seq.loc[i+1, 'predict'] = predict_1day
                model_seq.loc[i+1, 'lower_bound'] = predict_1day - (conf_int * np.std(resids[-rolling_window:]))
                model_seq.loc[i+1, 'upper_bound'] = predict_1day + (conf_int * np.std(resids[-rolling_window:]))

            else: 
                print('Anomaly detected, but trigger is NOT reached.')
                print(f'Anomalies in a row = {anomalies_in_a_row}')

                predict_1day = model.forecast(anomalies_in_a_row + 1)[-1]
                model_seq.loc[i+1, 'predict'] = predict_1day
                model_seq.loc[i+1, 'lower_bound'] = predict_1day - (conf_int * np.std(resids[-rolling_window:]))
                model_seq.loc[i+1, 'upper_bound'] = predict_1day + (conf_int * np.std(resids[-rolling_window:]))



        else: 
            model_seq.loc[i,'anomaly'] = 0
            anomalies_in_a_row = 0 

            resids = np.append(resids, predicted - observed)
            train_s = np.append(train_s, observed)

            model_seq.loc[i, 'resid'] = predicted - observed

            if len(train_s) >= 14: 
                model = ExponentialSmoothing(train_s[-rolling_window:], trend='add',
                                     seasonal='add',
                                     seasonal_periods=7,
                                     damped_trend=0.8).fit()
            else: 
                model = ExponentialSmoothing(train_s[-rolling_window:], trend='add', 
                                            initialization_method = 'estimated', 
                                            damped_trend = 0.3).fit()

            predict_1day = model.forecast(1)[0]

            model_seq.loc[i+1, 'predict'] = predict_1day
            model_seq.loc[i+1, 'lower_bound'] = predict_1day - (conf_int * np.std(resids[-rolling_window:]))
            model_seq.loc[i+1, 'upper_bound'] = predict_1day + (conf_int * np.std(resids[-rolling_window:]))


    model_seq = model_seq[:-1]   
    return model_seq
```

```python
def calc_errors(result):  
    real = result[start_index:]['real']
    pred = result[start_index:]['predict']
    print('MAE = ', mean_absolute_error(real, pred))
    print('R2 = ', r2_score(real, pred))
```
